## Analysis of Game Tags

```python
import tag_analysis
```

Any user who purchases a Steam game can select any number of 'tags' to associate with the game. Since these tags can be defined by the user, we see a much larger variety in tags than we do in 'genres' or 'categories,' which are defined by the developer.

First, we generate a list with every unique tag:

```python
tags = tag_analysis.get_tags()
```

Since tags are user-assigned, there are often tags that are incorrectly assigned to a game or that most users would not agree represents the game's content. Another issue is that popular games can receive thousands of tags while others do not, which could skew analysis towards tags associated with those popular games.

In order to solve these problems, we create a modified DataFrame in which for each game we assign a '1' to tags that appear very frequently and a '0' to the 'outlier' tags. 

Note that 'frequent' tags are defined here as being those that appear with more than half the frequency of the game's most popular tag.

```python
all_df = tag_analysis.generate_df()
```

### Change in Tags Over Time

First we determine the 8 tags that have had the largest amount of change in representation over the past 10 years:

```python
changed_tags, changed_ratios = tag_analysis.find_changed_tags_ratios(all_df, 8)
print(changed_tags)
```

Now let's see how these tag's prevalence has changed over time:

```python
for n in range(2):
    tag_analysis.plot_tag_prevalence(changed_tags[n*4:n*4 + 4], changed_ratios)
```

We can also explore the how the score distribution for specific tags changes throughout the years:

```python
tag_analysis.plot_tag_scores_per_year(all_df, 'indie')
tag_analysis.plot_tag_scores_per_year(all_df, 'third_person')
```

Note how the changes in scores reflect the relative prevalence of the indie and third-person genres throughout time.
