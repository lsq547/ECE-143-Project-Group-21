### Developers vs Publishers

- Developers are responsible for design and create games. They often specialize in 1-2 genres.
- Publishers are responsible for distribute, advertise and oversight the production of games. They release games in a variety of genres.

### Publishing History of Valve and Ubisoft

Valve has a unique publishing pattern. It has been a big producer of PC games since 90s and focused exclusively on action/FPS.
After Steam launched in 2003, Valve slowly decreased published output until stopped completely in 2013.
It is unusual for big publishers to focus on 1-2 genres

Ubisoft, on the other hand, shows a more common publishing pattern among top publishers. In 2008, Ubisoft almost exclusively focused on action and FPS games. However, in the following years there is an increasing diversity of genres, including new focus on adventure and RPG titles.
